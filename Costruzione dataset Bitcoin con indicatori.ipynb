```python
# Scarica i dati storici dei prezzi dei Bitcoin
dati = pd.read_csv('Bitcoin (€) da yfinance dal 17-09-2014 al 26-03-2024.csv')
dati
```

```python
dati = dati[['Date', 'Open', 'High', 'Low', 'Close']]
dati
```

```python
# Aggiungo varie colonne con le medie mobili sia semplici che esponenziali con varie lunghezze
for lunghezza in range(2, 201):
    dati.ta.sma(length=lunghezza, append=True)

for lunghezza in range(2, 201):
    dati.ta.ema(length=lunghezza, append=True) # il calcolo che viene fatto per la media mobile esponenziale è EMA = (Close - EMA_precedente) * 2/(lunghezza+1) + EMA_precedente
```

```python
dati_arricchiti = dati.dropna().reset_index(drop=True)
dati_arricchiti
```

### Costruisco il dataset di addestramento su cui farò machine learning

```python
# Numero di giorni nel futuro da prevedere
futuro = 7

# Crea un nuovo DataFrame per i dati di addestramento
dati_addestramento = pd.DataFrame()

# Aggiungi le colonne delle features al DataFrame di addestramento
for i in range(len(dati_arricchiti) - futuro):
    features = dati_arricchiti.iloc[i, 1:].tolist()
    dati_addestramento = dati_addestramento.append([features], ignore_index=True)

# Rinomina le colonne delle features
dati_addestramento.columns = dati_arricchiti.columns[1:]

# Crea un nuovo DataFrame per i target
dati_target = pd.DataFrame()

# Aggiungi le colonne dei target al DataFrame dei target
for i in range(1, len(dati_arricchiti) - futuro + 1):
    target = dati_arricchiti['Close'].iloc[i:i+futuro].tolist()
    dati_target = dati_target.append([target], ignore_index=True)

# Rinomina le colonne dei target
dati_target.columns = [f'Futuro_{j+1}' for j in range(futuro)]

# Unisci i DataFrame delle features e dei target
dati_finali = pd.concat([dati_addestramento, dati_target], axis=1)

# Stampa i primi dati per controllare
dati_finali
```

#### Aggiungo anche le variazioni dei prezzi per ogni giorno futuro

```python
dati_finali['Variazione futuro 1'] = np.where(dati_finali['Futuro_1'] - dati_finali['Close'] >= 0, "+", "-")
for i in range(2, futuro + 1):
    dati_finali[f'Variazione futuro {i}'] = np.where(dati_finali[f'Futuro_{i}'] - dati_finali[f'Futuro_{i-1}'] >= 0, "+", "-")
```
