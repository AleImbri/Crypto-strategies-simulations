```python
df = pd.read_csv(percorso_dataset, sep=",")
df
```

```python
df.isna().sum()
```

```python
df_senza_nan = df.dropna(inplace=False)
df_senza_nan
```

```python
# Ricavo la porzione di dataframe con il numero massimo di indici consecutivi

start = df_senza_nan.index[0]
i_start = 0
max_len = 0
max_range = (start, start)

for i in range(1, len(df_senza_nan.index)):
    if df_senza_nan.index[i] != df_senza_nan.index[i-1] + 1:
        if i - i_start > max_len:
            max_len = i - i_start
            max_range = (start, df_senza_nan.index[i-1])
        start = df_senza_nan.index[i]
        i_start = i

# Controlla l'ultimo intervallo
if len(df_senza_nan) - start > max_len:
    max_range = (start, df_senza_nan.index[-1])

# Stampa la porzione del DataFrame con il maggior numero di indici consecutivi
df_senza_nan.loc[max_range[0]:max_range[1]]
```
